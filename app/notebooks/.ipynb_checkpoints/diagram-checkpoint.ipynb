```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pymongo import MongoClient
import os
from wordcloud import WordCloud

# Thit lp font ch cho ting Vit
plt.rcParams['font.family'] = 'DejaVu Sans'

# Kt ni MongoDB
import os
import pymongo
from pymongo import MongoClient
mongo_uri = os.environ.get('MONGO_URI')
client = MongoClient(mongo_uri)
db = client.get_database()
collection = db['pred_news']

# Ly d liu t collection
data = list(collection.find())
df = pd.DataFrame(data)
# Kim tra d liu
print(f"Total number of records: {len(df)}")
print(df.columns.tolist())

# Tnh ton cc thng k v sentiment
sentiment_counts = df['sentiment'].value_counts()
pred_sentiment_counts = df['pred_sentiment'].value_counts()
comparison_df = pd.DataFrame({
    'Actual': sentiment_counts, 
    'Predicted': pred_sentiment_counts
})

# To confusion matrix
confusion = pd.crosstab(df['sentiment'], df['pred_sentiment'], 
                         rownames=['Actual'], colnames=['Predicted'])

# t style cho biu 
plt.style.use('ggplot')
sns.set(font_scale=1.2)

# To figure vi 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("Phn tch cm xc tin tc chng khon Vit Nam", fontsize=16)

# 1. Biu  ct th hin s lng ca tng loi sentiment
comparison_df.plot(kind='bar', ax=axes[0, 0], color=['#3498db', '#e74c3c'])
axes[0, 0].set_title('So snh s lng tht v d on')
axes[0, 0].set_ylabel('S lng')
axes[0, 0].set_xlabel('Loi cm xc')
axes[0, 0].legend(['Thc t', 'D on'])

# 2. Biu  trn cho phn b thc t
colors = ['#3498db', '#e74c3c', '#2ecc71'][:len(sentiment_counts)]
axes[0, 1].pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', 
              colors=colors, startangle=90)
axes[0, 1].set_title('Phn b cm xc thc t')

# 3. Confusion matrix
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', ax=axes[1, 0])
axes[1, 0].set_title('Ma trn nhm ln')

# 4.  chnh xc ca m hnh
accuracy = (df['sentiment'] == df['pred_sentiment']).mean() * 100
axes[1, 1].bar([' chnh xc'], [accuracy], color='#9b59b6')
axes[1, 1].set_ylim(0, 100)
axes[1, 1].set_title(' chnh xc ca m hnh')
axes[1, 1].set_ylabel('Phn trm (%)')
axes[1, 1].text(0, accuracy + 2, f'{accuracy:.1f}%', ha='center')

# Chnh layout
plt.tight_layout()
plt.subplots_adjust(top=0.9)

# Hin th  th
plt.show()

# To biu  ph: word cloud cho t kha ph bin trong tiu 
plt.figure(figsize=(12, 10))

# Kim tra xem ct 'title' c tn ti khng
if 'title' in df.columns:
    # X l text  to word cloud
    text = ' '.join(df['title'].astype(str).fillna(''))
    # Loi b cc t dng trong ting Vit
    vietnamese_stopwords = ['v', 'ca', 'c', 'cho', 'cc', 'vi', 'l', 'c', 'trong', 
                          '', 'ti', 't', 'theo', 'nhng', '', 'khng', 'ny', 'n']
    words = ' '.join([word.lower() for word in text.split() 
                     if word.lower() not in vietnamese_stopwords])

    # To word cloud
    wordcloud = WordCloud(width=800, height=500, 
                         background_color='white', 
                         max_words=100,
                         font_path='/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf').generate(words)

    # Hin th word cloud
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud - T kha ph bin trong tiu  tin tc', fontsize=16)
    plt.tight_layout()
    plt.show()
else:
    print("Khng tm thy ct 'title' trong d liu")
```
