```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pymongo import MongoClient
import os
from wordcloud import WordCloud

# Thiết lập font chữ cho tiếng Việt
plt.rcParams['font.family'] = 'DejaVu Sans'

# Kết nối MongoDB
import os
import pymongo
from pymongo import MongoClient
mongo_uri = os.environ.get('MONGO_URI')
client = MongoClient(mongo_uri)
db = client.get_database()
collection = db['pred_news']

# Lấy dữ liệu từ collection
data = list(collection.find())
df = pd.DataFrame(data)
# Kiểm tra dữ liệu
print(f"Total number of records: {len(df)}")
print(df.columns.tolist())

# Tính toán các thống kê về sentiment
sentiment_counts = df['sentiment'].value_counts()
pred_sentiment_counts = df['pred_sentiment'].value_counts()
comparison_df = pd.DataFrame({
    'Actual': sentiment_counts, 
    'Predicted': pred_sentiment_counts
})

# Tạo confusion matrix
confusion = pd.crosstab(df['sentiment'], df['pred_sentiment'], 
                         rownames=['Actual'], colnames=['Predicted'])

# Đặt style cho biểu đồ
plt.style.use('ggplot')
sns.set(font_scale=1.2)

# Tạo figure với 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("Phân tích cảm xúc tin tức chứng khoán Việt Nam", fontsize=16)

# 1. Biểu đồ cột thể hiện số lượng của từng loại sentiment
comparison_df.plot(kind='bar', ax=axes[0, 0], color=['#3498db', '#e74c3c'])
axes[0, 0].set_title('So sánh số lượng thật và dự đoán')
axes[0, 0].set_ylabel('Số lượng')
axes[0, 0].set_xlabel('Loại cảm xúc')
axes[0, 0].legend(['Thực tế', 'Dự đoán'])

# 2. Biểu đồ tròn cho phân bố thực tế
colors = ['#3498db', '#e74c3c', '#2ecc71'][:len(sentiment_counts)]
axes[0, 1].pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', 
              colors=colors, startangle=90)
axes[0, 1].set_title('Phân bố cảm xúc thực tế')

# 3. Confusion matrix
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', ax=axes[1, 0])
axes[1, 0].set_title('Ma trận nhầm lẫn')

# 4. Độ chính xác của mô hình
accuracy = (df['sentiment'] == df['pred_sentiment']).mean() * 100
axes[1, 1].bar(['Độ chính xác'], [accuracy], color='#9b59b6')
axes[1, 1].set_ylim(0, 100)
axes[1, 1].set_title('Độ chính xác của mô hình')
axes[1, 1].set_ylabel('Phần trăm (%)')
axes[1, 1].text(0, accuracy + 2, f'{accuracy:.1f}%', ha='center')

# Chỉnh layout
plt.tight_layout()
plt.subplots_adjust(top=0.9)

# Hiển thị đồ thị
plt.show()

# Tạo biểu đồ phụ: word cloud cho từ khóa phổ biến trong tiêu đề
plt.figure(figsize=(12, 10))

# Kiểm tra xem cột 'title' có tồn tại không
if 'title' in df.columns:
    # Xử lý text để tạo word cloud
    text = ' '.join(df['title'].astype(str).fillna(''))
    # Loại bỏ các từ dừng trong tiếng Việt
    vietnamese_stopwords = ['và', 'của', 'có', 'cho', 'các', 'với', 'là', 'được', 'trong', 
                          'đã', 'tại', 'từ', 'theo', 'những', 'để', 'không', 'này', 'đến']
    words = ' '.join([word.lower() for word in text.split() 
                     if word.lower() not in vietnamese_stopwords])

    # Tạo word cloud
    wordcloud = WordCloud(width=800, height=500, 
                         background_color='white', 
                         max_words=100,
                         font_path='/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf').generate(words)

    # Hiển thị word cloud
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Word Cloud - Từ khóa phổ biến trong tiêu đề tin tức', fontsize=16)
    plt.tight_layout()
    plt.show()
else:
    print("Không tìm thấy cột 'title' trong dữ liệu")
```
